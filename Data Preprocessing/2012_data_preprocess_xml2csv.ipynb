```python
from lxml.html import parse
from bs4 import BeautifulSoup
import csv
import os

threshold = 20

soup = BeautifulSoup(open("/Users/HUANGWEIJIE/Desktop/sogou_news_chinese_corpus/2012/tensite_ac"), "lxml")

with open(os.path.join('new_CSV_Data', 'tensite_afterProcess.csv'), 'a') as fhandle:
    writer = csv.writer(fhandle)
    writer.writerow(('classes', 'content'))
    for doc in soup.find_all('doc'):
        # Do Something there
        label = ""
        if "sports." in doc.url.text:
            label = "0"
        elif "business." in doc.url.text or "finance." in doc.url.text or "money." in doc.url.text:
            label = "1"
        elif "yule." in doc.url.text or "ent." in doc.url.text:
            label = "2"
        elif "auto."in doc.url.text:
            label = "3"
        elif "it." in doc.url.text or "tech." in doc.url.text:
            label = "4"
        else:
            continue
        
        content = doc.contenttitle.text + doc.content.text
        # Content太短的 直接去掉
        if len(content) <= threshold:
            continue
            
        writer.writerow((label, content))
```
